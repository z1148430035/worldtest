```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from pathlib import Path  
import pandas as pd
from pandas import Series
```

```python
poss = []
negs = []
alldata=[]
```

```python
df = pd.read_csv('D:/wordpython/data/tmo.csv',header=None,names=['Comment'])
```

```python
df
```

```python
for manner in df['Comment']:
    # print(manner)
    blob = TextBlob(manner, analyzer=NaiveBayesAnalyzer())
    data = blob.sentiment
    pos = round(data[1],2)
    poss.append(pos)
    # print(f"正向POS：{pos}")
    neg =round(data[2],2)
    # print(f"负面neg:{neg}")
    negs.append(neg)
    # manner(pos)
```

```python
ps = Series(poss,name="poss")
ng = Series(negs,name = "negs")
df['poss'] = poss
df['negs'] = negs
```

```python
data1=[]
data2=[]
for pos in df['poss']:
    if pos>0.5:
        data1.append(pos)
        data2.append('positive')
    elif pos == 0.5:
        data1.append(0.5)
        data2.append('neutral')
    else:
        data1.append(1-pos)
        data2.append('negative')
```

```python
df['Sentiment'] = data1
df['Appraise'] = data2
```

```python
df.to_csv('D:/wordpython/report/tmo_report.csv')
```
